```python
import pandas as pd

with open("./performance-comparison-table", "r") as f:
    df = pd.read_csv(f, delim_whitespace=True, engine='python')
df.head()
```

# The performance improvement of HUNTER over other PM file systems in Fileserver. 

```python
fs_df = df[df['file_bench'] == 'fileserver.f']

fs_names = fs_df['file_system'].drop_duplicates().reset_index().drop(columns=['index'])['file_system']

hunter =  fs_df[fs_df["file_system"] == "HUNTER"].copy().reset_index()

for fs_name in fs_names:
    if fs_name == "HUNTER":
        continue
    fs = fs_df[fs_df["file_system"] == fs_name]
    hunter[fs_name] = hunter["iops"] / fs["iops"].reset_index()["iops"]

hunter
```

# The performance improvement of HUNTER over other PM file systems in Varmail. 

```python
fs_df = df[df['file_bench'] == 'varmail.f']

fs_names = fs_df['file_system'].drop_duplicates().reset_index().drop(columns=['index'])['file_system']

hunter =  fs_df[fs_df["file_system"] == "HUNTER"].copy().reset_index()

for fs_name in fs_names:
    if fs_name == "HUNTER":
        continue
    fs = fs_df[fs_df["file_system"] == fs_name]
    hunter[fs_name] = hunter["iops"] / fs["iops"].reset_index()["iops"]

hunter
```

# The performance improvement of HUNTER-OP-ASYNC over other PM file systems in Varmail. 

```python
fs_df = df[df['file_bench'] == 'varmail.f']

fs_names = fs_df['file_system'].drop_duplicates().reset_index().drop(columns=['index'])['file_system']

hunter_op_async =  fs_df[fs_df["file_system"] == "HUNTER-NOSYNC"].copy().reset_index()

for fs_name in fs_names:
    if fs_name == "HUNTER-NOSYNC":
        continue
    fs = fs_df[fs_df["file_system"] == fs_name]
    hunter_op_async[fs_name] = hunter_op_async["iops"] / fs["iops"].reset_index()["iops"]
    
print("Compared to NOVA: "+str(hunter_op_async["NOVA"].mean()))
print("Compared to PMFS: "+str(hunter_op_async["PMFS"].mean()))
```
