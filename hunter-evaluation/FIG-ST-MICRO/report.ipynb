# Let's calc how much performance of PM are wasted

```python
# ipython -c "%run plot.ipynb"

import pandas as pd
import numpy as np


# Data Preprocessing
# with open("./avg-test-result-in-paper", "r") as f:
with open("./performance-comparison-table", "r") as f:
    df = pd.read_csv(f, delim_whitespace=True, index_col=None, engine='python')
with open("./performance-comparison-table-interleaved", "r") as f:
    df_int = pd.read_csv(f, delim_whitespace=True, index_col=None, engine='python')
with open("./performance-comparison-table-fsync", "r") as f:
    df_fsync = pd.read_csv(f, delim_whitespace=True, index_col=None, engine='python')
with open("./performance-comparison-table-fsync-interleaved", "r") as f:
    df_fsync_int = pd.read_csv(f, delim_whitespace=True, index_col=None, engine='python')

df.drop(df[df['file_system'] == "PMM"].index, inplace=True)

fss = df['file_system'].drop_duplicates().reset_index().drop(columns=['index'])['file_system']
ops = ["seq-write", "seq-write", "seq-write", "seq-write" ]


# gs0 = gridspec.GridSpec(2, 2, figure=fig, wspace=0.3, hspace=0.6)
wasted_percentage = []
cmap = "tab10"
for op_idx, op in enumerate(ops):
    if op_idx == 1:
        df = df_fsync
    if op_idx == 2:
        df = df_int
    if op_idx == 3:
        df = df_fsync_int
    # subfig = plt.subplot(gs0[op_idx])
    for idx, fs in enumerate(fss):
        if fs == "PMM" or fs == "HUNTER" or fs == "HUNTER-J" or fs == "SplitFS-FIO" or fs == "SoupFS":
            continue
        if df is df_int or df is df_fsync_int:
            PM_BW = 3336
        else:
            PM_BW = 2259.2
        wasted_percentage.append(1 - df[(df['file_system'] == fs) & (df['ops'] == op)]['bandwidth(MiB/s)'] / PM_BW)
print(wasted_percentage)
print(np.min(wasted_percentage))
print(np.max(wasted_percentage))
```
