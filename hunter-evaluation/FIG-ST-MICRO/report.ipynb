# Let's calc how much performance of PM are wasted

```python
# ipython -c "%run plot.ipynb"

import pandas as pd
import numpy as np


# Data Preprocessing
# with open("./avg-test-result-in-paper", "r") as f:
with open("./performance-comparison-table", "r") as f:
    df = pd.read_csv(f, delim_whitespace=True, index_col=None, engine='python')

df.drop(df[df['file_system'] == "PMM"].index, inplace=True)

fss = df['file_system'].drop_duplicates().reset_index().drop(columns=['index'])['file_system']
ops = ["seq-write", "rnd-write", ]
fsize = [1024, 2048, 4096, 8192, 12288, 16384, 20480, 24576, 28672, 32768]


# gs0 = gridspec.GridSpec(2, 2, figure=fig, wspace=0.3, hspace=0.6)
wasted_percentage = []
cmap = "tab10"
for op_idx, op in enumerate(ops):
    # subfig = plt.subplot(gs0[op_idx])
    for idx, fs in enumerate(fss):
        if fs == "PMM" or fs == "KILLER" or fs == "SplitFS-FIO":
            continue
        wasted_percentage.append(1 - df[(df['file_system'] == fs) & (df['ops'] == op)]['bandwidth(MiB/s)'] / 2259.2)
print(wasted_percentage)
print(np.min(wasted_percentage))
print(np.max(wasted_percentage))
```

# Let's see how much performance of HUNTER than others 

```python
# ipython -c "%run plot.ipynb"

import pandas as pd
import numpy as np


# Data Preprocessing
# with open("./avg-test-result-in-paper", "r") as f:
with open("./performance-comparison-table", "r") as f:
    df = pd.read_csv(f, delim_whitespace=True, index_col=None, engine='python')

df.drop(df[df['file_system'] == "PMM"].index, inplace=True)

fss = df['file_system'].drop_duplicates().reset_index().drop(columns=['index'])['file_system']
ops = ["seq-write" ]
fsize = [1024, 2048, 4096, 8192, 16384, 32768, 65536]


# gs0 = gridspec.GridSpec(2, 2, figure=fig, wspace=0.3, hspace=0.6)
improvement = []
cmap = "tab10"
for op_idx, op in enumerate(ops):
    # subfig = plt.subplot(gs0[op_idx])
    for idx, fs in enumerate(fss):
        if fs == "PMM" or fs == "HUNTER-J" or fs == "HUNTER":
            continue
        improvement.append(df[(df['file_system'] == "HUNTER-J") & (df['ops'] == op)]['bandwidth(MiB/s)'].reset_index().drop(columns=['index']) / df[(df['file_system'] == fs) & (df['ops'] == op)]['bandwidth(MiB/s)'].reset_index().drop(columns=['index']))
print(improvement)
print(np.min(improvement))
print(np.max(improvement))
```
